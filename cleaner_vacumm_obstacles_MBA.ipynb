```python
import random

# Cấu hình môi trường
N, M = 5, 5
environment = {}
for x in range(N):
    for y in range(M):
        # 0: Sạch, 1: Bẩn, 2: Vật cản
        environment[(x, y)] = random.choices([0, 1, 2], weights=[0.6, 0.3, 0.1])[0]
position = (0, 0)
environment[position] = random.choice([0, 1])
# Bộ nhớ lưu trữ thông tin các ô đã ghé thăm, số lần 
internal_model = {
    "visited": {},  
}
def update_model(pos):
    if pos not in internal_model["visited"]:
        internal_model["visited"][pos] = 1
    else:
        internal_model["visited"][pos] += 1

def interpret_input(percept):
    x, y = percept
    state_value = environment[(x, y)]
    update_model(percept)
    possible_moves = []
    # Kiểm tra các hướng (Up, Down, Left, Right) và tránh vật cản (2)
    directions = [("Up", (x-1, y)), ("Down", (x+1, y)), 
                  ("Left", (x, y-1)), ("Right", (x, y+1))]
    for move_name, (nx, ny) in directions:
        if 0 <= nx < N and 0 <= ny < M and environment[(nx, ny)] != 2:
            visit_count = internal_model["visited"].get((nx, ny), 0)
            possible_moves.append((move_name, visit_count))    
    return (percept, state_value, possible_moves)

def model_based_agent(percept):
    state = interpret_input(percept)
    pos, val, moves = state
    # bẩn thì hút
    if val == 1:
        return "SUCK"
    if not moves:
        return "STAY"
    # Ưu tiên ô có số lần thăm (visit_count) thấp nhất
    # Sắp xếp các hướng đi theo số lần đã thăm tăng dần
    moves.sort(key=lambda x: x[1])
    # Lấy danh sách các ô "ít được thăm nhất" (để nếu có nhiều ô cùng 0 lần thì chọn ngẫu nhiên)
    min_visits = moves[0][1]
    best_moves = [m[0] for m in moves if m[1] == min_visits]
    return random.choice(best_moves)

def main():
    global position
    print("--- BẢN ĐỒ MÔI TRƯỜNG (0:Sạch, 1:Bẩn, 2:Vật cản) ---")
    for r in range(N):
        print([environment[(r, c)] for c in range(M)])
    print("-" * 50)
    for step in range(20):
        x, y = position
        action = model_based_agent(position)
        print(f"Bước {step}: Vị trí={position}, Ô hiện tại={environment[(x,y)]}, Hành động={action}")
        if action == "SUCK":
            environment[(x, y)] = 0
        elif action == "Up": position = (x - 1, y)
        elif action == "Down": position = (x + 1, y)
        elif action == "Left": position = (x, y - 1)
        elif action == "Right": position = (x, y + 1)
    print("-" * 50)
    print("Bộ nhớ robot đã lưu (Số lần thăm các ô):")
    print(internal_model["visited"])

if __name__ == "__main__":
    main()
```
