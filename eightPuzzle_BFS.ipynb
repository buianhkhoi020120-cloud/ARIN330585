```python
from collections import deque

# Trạng thái đích của bàn cờ
GOAL = (
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 0)
)
class PuzzleNode:
    def __init__(self, state, parent=None, action=None):
        self.state = state      # Trạng thái hiện tại (tuple 3x3)
        self.parent = parent    # Nút cha để truy vết đường đi
        self.action = action    # Hành động dẫn đến trạng thái này
    def find_blank(self):
        for r in range(3):
            for c in range(3):
                if self.state[r][c] == 0:
                    return r, c
    # Tạo ra các trạng thái con hợp lệ bằng cách di chuyển ô trống
    def get_neighbors(self):
        neighbors = []
        r, c = self.find_blank()
        moves = [
            (-1, 0, 'UP'),   
            (1, 0, 'DOWN'), 
            (0, -1, 'LEFT'),  
            (0, 1, 'RIGHT')  
        ]
        for dr, dc, action in moves:
            nr, nc = r + dr, c + dc
            if 0 <= nr < 3 and 0 <= nc < 3:
                new_state = [list(row) for row in self.state]
                new_state[r][c], new_state[nr][nc] = new_state[nr][nc], new_state[r][c]
                tuple_state = tuple(tuple(row) for row in new_state)
                neighbors.append(PuzzleNode(tuple_state, self, action))           
        return neighbors
    
# Thuật toán BFS
def solve_bfs(start_state):
    # Khởi tạo nút gốc
    start_node = PuzzleNode(start_state)
    if start_node.state == GOAL:
        return start_node
    # Hàng đợi FIFO cho BFS và Tập các trạng thái đã duyệt
    queue = deque([start_node])
    visited = {start_node.state}
    while queue:
        current_node = queue.popleft()
        if current_node.state == GOAL:
            return current_node
        # Duyệt các trạng thái
        for neighbor in current_node.get_neighbors():
            if neighbor.state not in visited:
                visited.add(neighbor.state)
                queue.append(neighbor)  
    return None # Không tìm thấy lời giải

def print_matrix(state):
    for row in state:
        print(" ".join(str(x) for x in row))
    print("-" * 10)

def print_solution(solution_node):
    path = []
    current = solution_node
    while current:
        path.append(current)
        current = current.parent 
    path.reverse()
    for i, node in enumerate(path):
        if i == 0:
            print("Trạng thái ban đầu:")
        else:
            print(f"Bước {i}: {node.action}")
        print_matrix(node.state)

if __name__ == "__main__":
    initial_state = (
        (1, 2, 3),
        (4, 0, 6),
        (7, 5, 8)
    )
    print()
    result = solve_bfs(initial_state)
    if result:
        print_solution(result)
    else:
        print("Không tìm thấy lời giải cho trạng thái này!")
```
