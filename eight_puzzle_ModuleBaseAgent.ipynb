```python
goal = (1,2,3,4,5,6,7,8,0) #trạng thái đích
moves = {
    "UP": (-1,0),
    "DOWN": (1,0),
    "LEFT": (0,-1),
    "RIGHT": (0,1)
}
def available_moves(state):
    zero_pos = state.index(0)
    row, col = zero_pos // 3, zero_pos % 3
    possible_moves = []
    for move, (r, c) in moves.items():
        new_row, new_col = row + r, col + c
        if 0 <= new_col < 3 and 0 <= new_row < 3:
            possible_moves.append(move)
    return possible_moves
def print_matrix(state):
    for i in range(0, 9, 3):
        print(state[i: i + 3])
    print("-" * 10)
def is_solvable(state):
    inversions = 0
    arr = [x for x in state if x != 0]
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] > arr[j]:
                inversions += 1
    return inversions % 2 == 0
def UPDATE_STATE(state, action):
    state = list(state)
    zero_pos = state.index(0)
    row, col = zero_pos // 3, zero_pos % 3
    r, c = moves[action]
    new_row, new_col = row + r, col + c
    new_index = new_row * 3 + new_col
    state[zero_pos], state[new_index] = (state[new_index], state[zero_pos])
    return tuple(state)
def RULE_MATCH(state, visited):
    posible_moves = available_moves(state)
    for action in posible_moves:
        next_state = UPDATE_STATE(state, action)
        if next_state not in visited:
            return action
    return None
def MODEL_BASED_REFLEX_AGENT(state, visited):
    action = RULE_MATCH(state, visited)
    return action
def solve_eight_puzzle(initial_state):
    if not is_solvable(initial_state):
        print("Trạng thái không thể giải được!")
        return
    queue = [(initial_state, [])]
    visited = set()
    while queue:
        current_state, path = queue.pop(0)
        if current_state in visited:
            continue
        visited.add(current_state)
        if current_state == goal:
            print("=== Ban đầu ===")
            print_matrix(initial_state)
            temp = initial_state
            print("=== Quá trình giải ===")
            for step, move in enumerate(path, start=1):
                temp = UPDATE_STATE(temp, move)
                print(f"Bước {step}: {move}")
                print_matrix(temp)
            print("=== Kết quả ===")
            print("Số bước:", len(path))
            print("Đường đi:", " -> ".join(path))
            return
        valid_moves = available_moves(current_state)
        for move in valid_moves:
            next_state = UPDATE_STATE(current_state, move)
            if next_state not in visited:
                queue.append((next_state, path + [move]))
if __name__ == "__main__":

    initial_puzzle = (
        1, 2, 3,
        6, 4, 5,
        7, 0, 8
    )
    solve_eight_puzzle(initial_puzzle)


```
