```python
import random

N = 3 
M = 4
environment = {}

# Khởi tạo môi trường với: 0 (Sạch), 1 (Bẩn), 2 (Vật cản)
for x in range(N):
    for y in range(M):
        # Tỉ lệ xuất hiện: 60% sạch, 25% bẩn, 15% vật cản
        environment[(x, y)] = random.choices([0, 1, 2], weights=[0.6, 0.25, 0.15])[0]

# Cố định Vị trí xuất phát (0, 0) không có vật cản
environment[(0, 0)] = random.choice([0, 1])
position = (0, 0)

# ===== INTERPRET INPUT =====
def interpret_input(percept):
    x, y = percept
    state_value = environment[(x, y)]
    moves = []

    if x > 0 and environment[(x - 1, y)] != 2:
        moves.append("Up")
    if x < N - 1 and environment[(x + 1, y)] != 2:
        moves.append("Down")
    if y > 0 and environment[(x, y - 1)] != 2:
        moves.append("Left")
    if y < M - 1 and environment[(x, y + 1)] != 2:
        moves.append("Right")

    state = (percept, state_value, moves)
    return state

def rule_match(state, rules):
    for rule in rules:
        if rule["condition"](state):
            return rule
    return None

rules = [
    {
        "condition": lambda state: state[1] == 1,  # Nếu ô hiện tại Bẩn
        "action": "SUCK"
    },
    {
        "condition": lambda state: state[1] == 0,  # Nếu ô hiện tại Sạch
        "action": "MOVE"
    }
]

# ===== SIMPLE REFLEX AGENT  =====
def simple_reflex_agent(percept):
    state = interpret_input(percept)
    rule = rule_match(state, rules)
    pos, val, moves = state

    if rule["action"] == "SUCK":
        return "SUCK"
    else:
        if not moves:
            return "STAY"
        return random.choice(moves)
    
def main():
    global position
    print("--- TRẠNG THÁI MÔI TRƯỜNG (0:Sạch, 1:Bẩn, 2:Vật cản) ---")
    for r in range(N):
        row = [environment[(r, c)] for c in range(M)]
        print(row)
    print("-" * 50)

    for step in range(20):
        x, y = position
        percept = position
        action = simple_reflex_agent(percept)
        
        print(f"Step {step}: Vị trí ={position}, Trạng thái = {environment[(x,y)]}, Hành động = {action}")

        if action == "SUCK":
            environment[(x, y)] = 0
        elif action == "Up":
            position = (x - 1, y)
        elif action == "Down":
            position = (x + 1, y)
        elif action == "Left":
            position = (x, y - 1)
        elif action == "Right":
            position = (x, y + 1)
        elif action == "STAY":
            pass # Đứng yên nếu bị kẹt

if __name__ == "__main__":
    main()
```
